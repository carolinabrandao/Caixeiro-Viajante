```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('results.csv')
```

```python
#count the number of times each algorithm was used
print(df['Algoritmo'].value_counts())
```

```python
#menor e maior qualidade de solucao dos dois algoritmos junto com o numero de n√≥s equivalente
print(df.groupby(['Algoritmo'])['Qualidade'].min())
print(df.groupby(['Algoritmo'])['Qualidade'].max())

```

```python
twice_around_tree = df[df['Algoritmo'] == 'Twice Around the Tree']
christofides = df[df['Algoritmo'] == 'Christofides']

# Configura√ß√£o do gr√°fico
plt.figure(figsize=(10, 6))

# Plotar os dados para Twice Around the Tree com linha
plt.plot(twice_around_tree['N√≥s'], twice_around_tree['Qualidade'], label='Twice Around the Tree', marker='o')

# Plotar os dados para Christofides com linha
plt.plot(christofides['N√≥s'], christofides['Qualidade'], label='Christofides', marker='x')

# Adicionar r√≥tulos e t√≠tulo
plt.xlabel('N√∫mero de N√≥s')
plt.ylabel('Qualidade')
plt.title('Qualidade dos Algoritmos para Diferentes N√∫meros de N√≥s')
plt.legend()  # Adicionar legenda

# Exibir o gr√°fico
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Leitura do CSV
df = pd.read_csv('results.csv')

# Filtrar os dados para Twice Around the Tree e Christofides
twice_around_tree = df[df['Algoritmo'] == 'Twice Around the Tree']
christofides = df[df['Algoritmo'] == 'Christofides']

# Configura√ß√£o do gr√°fico
plt.figure(figsize=(10, 6))

# Plotar os dados para Twice Around the Tree com linha
plt.plot(twice_around_tree['N√≥s'], twice_around_tree['Mem√≥ria(bytes)'], label='Twice Around the Tree', marker='o')

# Plotar os dados para Christofides com linha
plt.plot(christofides['N√≥s'], christofides['Mem√≥ria(bytes)'], label='Christofides', marker='x')

# Adicionar r√≥tulos e t√≠tulo
plt.xlabel('N√∫mero de N√≥s')
plt.ylabel('Mem√≥ria (bytes)')
plt.title('Consumo de Mem√≥ria em Fun√ß√£o do N√∫mero de N√≥s')
plt.legend()  # Adicionar legenda

# Exibir o gr√°fico
plt.tight_layout()
plt.show()
```
